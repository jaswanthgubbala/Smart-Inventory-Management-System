```python
import mysql.connector

# Database connection
db_connection = mysql.connector.connect(
    host="localhost",
    user="root",
    password="your password",
    database="inventory"
)

db_cursor = db_connection.cursor()


# ➤ Insert Product
def insert_product():
    pname = input("Enter product name: ")
    pprice = int(input("Enter price: "))
    pqty = int(input("Enter quantity: "))

    sql = "INSERT INTO products(name, price, quantity) VALUES(%s,%s,%s)"
    db_cursor.execute(sql, (pname, pprice, pqty))
    db_connection.commit()

    print("Product inserted successfully!")


# ➤ Display Products
def display_products():
    db_cursor.execute("SELECT * FROM products")
    records = db_cursor.fetchall()

    print("\nID | Name | Price | Quantity")
    print("--------------------------------")

    for item in records:
        print(item)

        # Low stock warning
        if item[3] < 10:
            print("⚠ Warning: Stock is low!")


# ➤ Modify Product
def modify_product():
    pid = int(input("Enter product ID: "))
    new_qty = int(input("Enter updated quantity: "))

    sql = "UPDATE products SET quantity=%s WHERE id=%s"
    db_cursor.execute(sql, (new_qty, pid))
    db_connection.commit()

    print("Product quantity updated!")


# ➤ Remove Product
def remove_product():
    pid = int(input("Enter product ID: "))

    sql = "DELETE FROM products WHERE id=%s"
    db_cursor.execute(sql, (pid,))
    db_connection.commit()

    print("Product removed successfully!")


# ➤ Process Sale
def process_sale():
    pid = int(input("Enter product ID: "))
    sell_qty = int(input("Enter quantity to sell: "))

    db_cursor.execute("SELECT quantity FROM products WHERE id=%s", (pid,))
    result = db_cursor.fetchone()

    if result:
        current_stock = result[0]

        if sell_qty <= current_stock:
            updated_stock = current_stock - sell_qty
            db_cursor.execute(
                "UPDATE products SET quantity=%s WHERE id=%s",
                (updated_stock, pid)
            )
            db_connection.commit()
            print("Sale completed successfully!")
        else:
            print("Insufficient stock available!")
    else:
        print("Invalid product ID!")


# ➤ Summary Dashboard
def show_dashboard():
    db_cursor.execute("SELECT COUNT(*) FROM products")
    total_items = db_cursor.fetchone()[0]

    db_cursor.execute("SELECT SUM(quantity) FROM products")
    total_quantity = db_cursor.fetchone()[0]

    print("\n--- Inventory Summary ---")
    print("Total Product Types:", total_items)
    print("Total Quantity Available:", total_quantity)


# ➤ Main Menu
while True:
    print("\n===== Stock Management System =====")
    print("1. Insert Product")
    print("2. Display Products")
    print("3. Modify Product")
    print("4. Remove Product")
    print("5. Process Sale")
    print("6. View Dashboard")
    print("7. Exit")

    option = int(input("Enter your choice: "))

    if option == 1:
        insert_product()
    elif option == 2:
        display_products()
    elif option == 3:
        modify_product()
    elif option == 4:
        remove_product()
    elif option == 5:
        process_sale()
    elif option == 6:
        show_dashboard()
    elif option == 7:
        print("Program closed. Thank you!")
        break
    else:
        print("Please enter a valid option!")
```
